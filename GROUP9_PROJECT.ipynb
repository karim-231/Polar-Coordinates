               A radar screen with direction controlled by tetha which lies within [0,2PI] and 
               how far towards the edge of the screen controlled by r which lies within  [1,5]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.subplot(111,projection = 'polar') 
r = list(range(1,6))
tetha = np.linspace(0,2*np.pi)
for r in r:
    x = r*np.cos(tetha)
    y = r*np.sin(tetha)
    plt.plot(tetha,[r]*len(tetha)) #### [r]*len(tetha) for each radius in r plot each value of tetha and r
plt.title('A giant radar screen')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Curve, r = 4cos +2 evolves with ')
for i in range(1,5):
    tetha1 = np.linspace(0,2*np.pi,lis[i-1]) #tetha1 = np.linspace(0,2*np.pi,50)
    plt.subplot(2,2,i,projection='polar')
    r1 = 4*np.cos(tetha1) + 2
    plt.plot(tetha1,r1,color='red')
    plt.title(f' with {lis[i-1]} points')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Cochleod Curve evolves with ')
for i in range(1,5):
    tetha2 = np.linspace(0.0001,2*np.pi,lis[i-1]) 
    plt.subplot(2,2,i,projection='polar')
    r2 = 3*np.sin(tetha2)/tetha2
    plt.plot(tetha2,r2,color='teal')
    plt.title(f' with {lis[i-1]} points')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Conchoid of de Sluze Curve evolves with ')
for i in range(1,5):
    tetha3 = np.linspace(0.01,2*np.pi,lis[i-1]) 
    plt.subplot(2,2,i,projection='polar')
    r3 = 2*np.cos(tetha3)/1-1/np.cos(tetha3)
    plt.plot(tetha3,r3,color='violet')
    plt.title(f' with {lis[i-1]} points')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Rhodonea Curve evolves with ')
for i in range(1,5):
    tetha4 = np.linspace(0,20*np.pi,lis[i-1]) 
    plt.subplot(2,2,i,projection='polar')
    r4 = np.sin(np.pi*tetha4)
    plt.plot(tetha4,r4,color='brown')
    plt.title(f' with {lis[i-1]} points')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Butterfly Curve evolves with ')
for i in range(1,5):
    tetha5 = np.linspace(0,2*np.pi,lis[i-1]) 
    plt.subplot(2,2,i,projection='polar')
    r5 = np.exp(np.cos(tetha5)) - 2*np.cos(4*tetha5) + np.sin((2*tetha5-np.pi)/25)**5
    plt.plot(tetha5,r5,color='blue')
    plt.title(f' with {lis[i-1]} points')
```

```python
lis = [50,100,500,1000]
plt.figure(figsize=(10,11))
plt.suptitle('How The Chrysanthemum Curve evolves with ')
for i in range(1,5):
    tetha6 = np.linspace(0,21*np.pi,lis[i-1]) 
    plt.subplot(2,2,i,projection='polar')
    r6=5*(1 + np.sin(11*tetha6/5))-(4*np.sin(17*tetha6/3)**4)*((np.sin(2*np.cos(3*tetha6) - 28*tetha6))**8)
    plt.plot(tetha6,r6,color='green')
    plt.title(f' with {lis[i-1]} points')
```
